```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
def calc_variance(file_names):

    img = cv2.imread(file_names[0])
    img_avg = np.zeros_like(img, dtype=np.float64)
    for name in file_names:
        img = cv2.imread(name)
        img_avg += img
    img_avg /= len(file_names)
    plt.imshow(cv2.cvtColor(img_avg.astype(np.uint8), cv2.COLOR_BGR2RGB))
    img_avg_gray = cv2.cvtColor(img_avg.astype(np.uint8), cv2.COLOR_BGR2GRAY)
    variance = np.zeros_like(img_avg_gray, dtype=np.float)
    for name in file_names:
        img = cv2.imread(name)
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float)
        variance += (img_gray - img_avg_gray) ** 2
    variance /= len(file_names)
    return variance
```

```python
variance_3 = calc_variance(glob.glob("../cam_3/*.jpg"))
plt.imshow(variance_3)
plt.colorbar()
plt.show()
```

```python
variance_0 = calc_variance(glob.glob("../cam_0/*.jpg"))
plt.imshow(variance_0)
plt.colorbar()
plt.show()
```

```python
threshold = 200
plt.subplot(121)
plt.imshow(variance_3<threshold)
plt.subplot(122)
plt.imshow(variance_0<threshold)
plt.show()
```
