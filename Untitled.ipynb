```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import glob
%matplotlib inline
```

```python
file_names = glob.glob("*.jpg")
img = cv2.imread(file_names[0])
img_avg = np.zeros_like(img, dtype=np.float64)
for name in file_names:
    img = cv2.imread(name)
    img_avg += img
img_avg /= len(file_names)
plt.imshow(cv2.cvtColor(img_avg.astype(np.uint8), cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(img_avg.astype(np.uint8), cv2.COLOR_BGR2GRAY), cmap="gray")
```

```python
img_avg_gray = cv2.cvtColor(img_avg.astype(np.uint8), cv2.COLOR_BGR2GRAY)
plt.imshow(img_avg_gray<100, cmap="gray")
plt.show()
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap="gray")
plt.show()
```
